```python
from population_generator import PopulationGenerator as pg
import numpy as np
import matplotlib.pyplot as plt
pops = pg.create()
#keep last four columns
pops = pops.iloc[:, -5:]
#to numpy
pops = pops.to_numpy()

pops[:20], type(pops), pops.shape
```

```python
class ClarvoyantAgent():
    def find_last_one(row):
        idx = np.where(row == 1)[0]

        return idx[-1] if len(idx) > 0 else 0

    def select_price(self, person):
        return ClarvoyantAgent.find_last_one(person)


```

```python
conversions = []
agent = ClarvoyantAgent()
for person in pops:
    proposoal = agent.select_price(person)
    conversion = 1 if person[proposoal] == 1 else 0
    conversions.append(conversion)

cumulative_conversions = np.cumsum(conversions)

print("Total conversions: {}".format(cumulative_conversions[-1]))
#plot cumulative conversions
plt.plot(cumulative_conversions)
plt.show()
```
